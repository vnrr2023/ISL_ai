```python
import numpy as np
```

```python
import os
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dir='../data/welcome/'
```

```python
# BlANK
```

```python
blank=[]
```

```python
for i in os.listdir(data_dir):
    temp=[]
    for folder in os.listdir(os.path.join(data_dir,i)):
        path=os.path.join(data_dir,i,folder)
        res=(np.load(path))
        temp.append(res)
    print(i)
    blank.append(temp)
    
    
```

```python
blank=np.array(blank)
```

```python
for i in range(30):
    sns.distplot((blank[0][i]),kde=True,color='purple')
    print(max(blank[0][i]))
    plt.show()
```

```python
sns.distplot(blank[0],color='purple')
plt.savefig("welcome.png")
# print(np.max(blank[0]))
# plt.show()
```
